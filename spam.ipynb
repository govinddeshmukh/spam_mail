```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/home/govind/Documents/pycharm-community-2021.2.1/programs/spam/spam_detection/spam.csv")
```

```python
df.head()
```

```python
df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis='columns', inplace=True)
```

```python
df
```

```python
df['v1'].replace('spam', 1, inplace=True)
df['v1'].replace('ham', 0, inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
x=df['v1']
y=df['v2']
```

```python
vectorizer = CountVectorizer()
```

```python
vectorizer.fit(y)
```

```python
print(vectorizer.vocabulary_)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(y, x, train_size=0.8, test_size=0.2, random_state=3)
```

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase='False')
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
model = LinearSVC()
model.fit(X_train_features, Y_train)

```

```python
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
print('Accuracy on test data : ', accuracy_on_test_data)
```

```python
input_mail = ["ree entry in 2 a wkly comp to win FA Cup"]
# convert text to feature vectors
input_mail_features = feature_extraction.transform(input_mail)

#making prediction
prediction = model.predict(input_mail_features)
print(prediction)

if (prediction[0]==1):
  print('HAM MAIL')
else:
  print('SPAM MAIL')
```
